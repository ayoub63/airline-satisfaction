Airline Customer Satisfaction Prediction

## 1. Introduction
This notebook explores machine learning models to predict airline passenger satisfaction.  
We compare Logistic Regression, Random Forest, XGBoost, and LightGBM.  
Finally, I use SHAP to interpret the models.

# 2. Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

import shap

```

## 3. Data Loading & Preprocessing

We start by loading the training and test data, cleaning missing values, and encoding categorical features.  
The target variable (`satisfaction`) is mapped to binary values (1 = satisfied, 0 = neutral or dissatisfied).  

Steps performed:
1. Load CSV files (`train.csv` and `test.csv`).
2. Drop unnecessary ID columns.
3. Fill missing values in *Arrival Delay in Minutes* with `0`.
4. Encode the target variable into binary format.
5. One-hot encode categorical variables (`Gender`, `Customer Type`, `Type of Travel`, `Class`).
6. Ensure train and test sets have aligned feature columns.

```python
TRAIN_PATH = "../data/train.csv"
TEST_PATH  = "../data/test.csv"

# Load CSVs
df_train = pd.read_csv(TRAIN_PATH)
df_test  = pd.read_csv(TEST_PATH)

# Drop ID Column
df_train = df_train.drop(columns=["Unnamed: 0", "id"])
df_test = df_test.drop(columns=["Unnamed: 0", "id"])

# Fill missing values in Arrival Delay
df_train["Arrival Delay in Minutes"] = df_train["Arrival Delay in Minutes"].fillna(0)
df_test["Arrival Delay in Minutes"]  = df_test["Arrival Delay in Minutes"].fillna(0)

# Encode Target Variable
df_train["satisfaction"] = df_train["satisfaction"].map({
    "satisfied": 1,
    "neutral or dissatisfied": 0
})

# Features and target
X_train_raw = df_train.drop(columns=["satisfaction"])
y_train = df_train["satisfaction"]

X_test_raw = df_test.drop(columns=["satisfaction"])
y_test = df_test["satisfaction"].map({
    "satisfied": 1,
    "neutral or dissatisfied": 0
})

# Identify categorical and numeric features
categorical_cols = ["Gender", "Customer Type", "Type of Travel", "Class"]
numeric_cols = [c for c in X_train_raw.columns if c not in categorical_cols]

#One-Hot Encode categoricals
X_train = pd.get_dummies(X_train_raw, columns=categorical_cols, drop_first=True)
X_test = pd.get_dummies(X_test_raw, columns=categorical_cols, drop_first=True)

# Align columns so train and test have the same set
X_train, X_test = X_train.align(X_test, join="left", axis=1, fill_value=0)
```

## 4. Model Training & Evaluation

I compare four different models:  
- Logistic Regression  
- Random Forest  
- XGBoost  
- LightGBM  

**Workflow**:  
1. Perform 5-fold Cross Validation on the training set.  
2. Train each model on the full training set.  
3. Evaluate performance on the hold-out test set.  
4. Collect results for accuracy, precision, recall, and F1-score. 

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

#  Define models 
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=300, max_depth=12, random_state=42, n_jobs=-1),
    "XGBoost": XGBClassifier(
        n_estimators=300,
        max_depth=8,
        learning_rate=0.1,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42,
        n_jobs=-1,
        use_label_encoder=False,
        eval_metric="logloss"
    ),
    "LightGBM": LGBMClassifier(
        n_estimators=300,
        max_depth=-1,
        learning_rate=0.1,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42,
        n_jobs=-1
    )
}

results = []

# Train + Cross Validation + Test 
for name, model in models.items():
    print(f"\nTraining {name}...")
    
    # Cross Validation on training set
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring="accuracy", n_jobs=-1)
    
    # Fit on full training data
    model.fit(X_train, y_train)
    
    # Evaluate on test set
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    prec = precision_score(y_test, preds)
    rec = recall_score(y_test, preds)
    f1 = f1_score(y_test, preds)
    
    results.append({
        "Model": name,
        "CV Mean": cv_scores.mean(),
        "CV Std": cv_scores.std(),
        "Test Accuracy": acc,
        "Test Precision": prec,
        "Test Recall": rec,
        "Test F1": f1
    })

# Results Table
results_df = pd.DataFrame(results).set_index("Model")
print(results_df.round(3))
```

### 5. Model Explainability with SHAP

To better understand **why** our models make certain predictions, We use **SHAP (SHapley Additive exPlanations)**.  
SHAP assigns each feature a contribution value that explains its impact on the modelâ€™s prediction.

- We focus on the **tree-based models** (LightGBM, XGBoost, Random Forest) since SHAP is optimized for them.  
- To speed up computation, we sample **1,000 rows from the test set**.  
- For binary classification, we analyze the SHAP values for the **positive class (Satisfied = 1)**.  

We generate two types of visualizations for each model:
1. **Global Importance (bar plot)**: Average absolute SHAP values show which features drive predictions the most.  
2. **Beeswarm Plot**: Shows the distribution and direction (positive/negative) of feature impact across many samples.

```python
import shap, numpy as np
import matplotlib.pyplot as plt


shap.explainers._tree.use_numba = False
shap.initjs()

# Use a subset for SHAP 
N_SHAP = 1000
X_shap = X_test.sample(min(N_SHAP, len(X_test)), random_state=42).copy()

def _shap_values(explainer, X):
    sv = explainer.shap_values(X)
    # For binary classifiers, TreeExplainer often returns [sv_class0, sv_class1]
    if isinstance(sv, list):
        sv = sv[1]  # focus on the "Satisfied=1" class
    return sv

for name in ["LightGBM", "XGBoost", "Random Forest"]:
    print(f"\n{name} â€” SHAP on {len(X_shap)} rows")
    model = models[name]
    # tree_path_dependent = fast approximation for tree models
    explainer = shap.TreeExplainer(
        model,
        feature_perturbation="tree_path_dependent",
        model_output="raw"
    )
    sv = _shap_values(explainer, X_shap)

    # Global (mean |SHAP|)
    shap.summary_plot(sv, X_shap, plot_type="bar", show=False, max_display=15)
    plt.title(f"{name} â€” SHAP (global importance)")
    plt.tight_layout()
    plt.show()

    # Beeswarm
    shap.summary_plot(sv, X_shap, show=False, max_display=20)
    plt.title(f"{name} â€” SHAP (beeswarm)")
    plt.tight_layout()
    plt.show()
```
